### Linear SVM Regression using the LinearSVR class
Recall:  
Instead of trying to get the widest street while minimizing the amt of data in the street like in SVM classification,  
SVM Regression tries to **fit as many instances as possible on the street while minimizing margin violations**

```python
import numpy as np
from sklearn.svm import LinearSVR

X = 20 * np.random.random((100, 1)) - 10
y = X + 4 + (4 * np.random.rand(100, 1))

svm_reg = LinearSVR(epsilon=1.5)
# Epsilon: In training, in the loss function, any predictions w/i epsilon distance of the actual value will have no penalty
svm_reg.fit(X, y.reshape(-1))
```

```python
X_validate = np.arange(-10, 10).reshape(20, 1)
y_validate = svm_reg.predict(X_validate)

import matplotlib.pyplot as plt
plt.plot(X, y, "b.")
plt.plot(X_validate, y_validate, "r-")
plt.show()
```

```python
svm_reg.coef_, svm_reg.intercept_ # predictions not bad
```

### Nonlinear SVM Regression using the SVR class

```python
from sklearn.svm import SVR

X = 20 * np.random.random((100, 1)) - 10
y = 3 * X ** 2 + 2 * X + 4 + (40 * np.random.rand(100, 1))

svm_poly_reg = SVR(kernel="poly", degree=2, C=100, epsilon=0.5)
# Epsilon: In training, in the loss function, any predictions w/i epsilon distance of the actual value will have no penalty
svm_poly_reg.fit(X, y.reshape(-1))
```

```python
X_validate = np.arange(-10, 10).reshape(20, 1)
y_validate = svm_poly_reg.predict(X_validate)

import matplotlib.pyplot as plt
plt.plot(X, y, "b.")
plt.plot(X_validate, y_validate, "r-")
plt.show()
```
